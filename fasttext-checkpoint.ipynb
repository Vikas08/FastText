```python
#import fasttext library
import fasttext
```

```python
#creating skipgram model
model = fasttext.skipgram("/home/nss/Documents/fasttext.txt",'model')
```

```python
print(model.words)
```

```python
#making cbow model
model_cbow = fasttext.cbow("/home/nss/Documents/fasttext.txt",'model')
```

```python
print(model_cbow.words)
```

```python
#word vector for 'king'
print(model['king'])
```

```python
print(model_cbow['king'])
```
